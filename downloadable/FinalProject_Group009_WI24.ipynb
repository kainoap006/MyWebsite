# Abstract

The 2023 NFL season experienced a unique phenomenon as popstar Taylor Swift debuted her new relationship with NFL player Travis Kelce. Throughout the season, her appearance at Kelce's games for the Kansas City Chiefs garnered increased public attention, which, in conjunction with heightened interest and hopes for the Chiefs to succeed and go to the Superbowl in 2024, contributed to the Chief's biggest year in viewership during the last decade. In this project, we hope to identify if there is a relation between Taylor Swift and Travis Kelce’s relationship and changes in engagement with the NFL and the Kansas City Chiefs with NFL TV viewership and Google Search trends over the last 10 years as our metrics for engagement. We are predicting that NFL TV viewership and Google search volume will increase for games in the 2023 season when Taylor is present; our null hypothesis is that there will be no significant change in TV viewership or search volume. We scraped web data on our chosen forms of engagement and created data visualizations and performed linear regression to gauge the statistical significance of Taylor Swift and Travis Kelce on Google searches for the Chiefs, where we found that there was only statistical significance for the effect of increased search volume for Taylor and Travis relative to searches for the Kansas City Chiefs. We also created ARIMA and SARIMA models to predict the expected TV viewership for the season when Taylor and Travis began dating (2023); SARIMA in particular showed that there is very little difference between what we expected in terms of viewership for 2023 and what the actual observed viewership for the 2023 season was, a result we confirmed using a Ljung-Box Test. Ultimately, we fail to reject the null hypothesis as we have very little evidence to indicate that there was a substantial relationship between Taylor Swift and Travis Kelce's relationship and changes in viewership and Google Search volume for the Chiefs and NFL.

Our data has 1206 observations across 8 different columns. Each particular row denotes a single game - note that our data is grouped, it would be difficult to get access to individual data so there has already been summary done such as rating/viewer numbers. Despite this fact, we can discuss the descriptions for each column:
 - Week represents what week each game was played on. We have to be careful because the NFL added an extra week into the regular season in 2021 meaning that seasons before 2021, week 21 would represent the superbowl, but not for seasons past 2021.
 - Network represents what broadcast network streamed the games and are what the viewership numbers are based off of.

```python
# Highlighting games with Taylor Swift's attendance
games_2023 = viewer_combined[(viewer_combined['Season'] == 2023) & ((viewer_combined['Team 1'] == 'KC') | (viewer_combined['Team 2'] == 'KC'))]
ts_games_2023 = viewer_combined[(viewer_combined['TS_Attendance'] == 1) &  ((viewer_combined['Team 1'] == 'KC') | (viewer_combined['Team 2'] == 'KC'))]
no_ts_games_2023 = viewer_combined[(viewer_combined['TS_Attendance'] == 0) & ((viewer_combined['Team 1'] == 'KC') | (viewer_combined['Team 2'] == 'KC')) & (viewer_combined['Season'] == 2023)]
no_ts_games_2022 = viewer_combined[(viewer_combined['TS_Attendance'] == 0) & ((viewer_combined['Team 1'] == 'KC') | (viewer_combined['Team 2'] == 'KC')) & (viewer_combined['Season'] == 2022)]
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.scatter(ts_games_2023['Week'], ts_games_2023['Viewers'], color='red', label='Taylor Swift Attendance (2023)')
ax1.scatter(no_ts_games_2023['Week'], no_ts_games_2023['Viewers'], color='blue', label='No Taylor Swift Attendance (2023)')
x1 = np.array(games_2023['Week'])
y1 = np.array(games_2023['Viewers'])
ax2.scatter(no_ts_games_2022['Week'], no_ts_games_2022['Viewers'], color='purple', label='No Taylor Swift Attendance (2022)')
x2 = np.array(no_ts_games_2022['Week'])
y2 = np.array(no_ts_games_2022['Viewers'])
b1, b0 = np.polyfit(x1, y1, 1)
a1, a0 = np.polyfit(x2, y2, 1) 
# Plot regression line
ax1.plot(x1, b1 * x1 + b0, color="k", lw=2.5)
ax2.plot(x2, a1 * x2 + a0, color="k", lw=2.5)
#Formatting
ax1.set_ylim(0, 60000000)
ax2.set_ylim(0, 60000000)
def millions_formatter(x, pos):
    return f'{x / 1000000}'
l1 = ax1.yaxis.set_major_formatter(FuncFormatter(millions_formatter))
l2 = ax2.yaxis.set_major_formatter(FuncFormatter(millions_formatter))
ax1.set_xlabel('Weeks')
ax2.set_xlabel('Weeks')
fig.suptitle('Kansas City Chiefs Viewership 2022-2023')
plt.figlegend(loc = 'upper right', ncol=1, bbox_to_anchor=(1.3, 1))
plt.subplots_adjust(right=1.2, hspace=0.25)
fig.text(0.04, 0.5, 'Viewers (in millions)', va='center', rotation='vertical');
```

```python
years = []
medianViewers = []
meanViewers = []
KC_games = viewer_combined[((viewer_combined['Team 1'] == 'KC') | (viewer_combined['Team 2'] == 'KC'))]
for year in KC_games.groupby("Season")["Viewers"].median().keys():
    years.append(year)
    medianViewers.append(KC_games.groupby("Season")["Viewers"].median()[year])
    meanViewers.append(KC_games.groupby("Season")["Viewers"].mean()[year])
fig, ax = plt.subplots(1,1)
plt.plot(years,medianViewers,label="Median Viewers")
plt.plot(years,meanViewers,label="Mean Viewers")
ax.yaxis.set_major_formatter(FuncFormatter(millions_formatter))
ax.set_ylim(10000000, 40000000)
fig.text(0.04, 0.5, 'Viewers (in millions)', va='center', rotation='vertical')
plt.legend();
```

We can see that our dataset consists of rows with 5 columns, including our 4 keywords used for searching: "Taylor Swift", "Travis Kelce", "NFL", and "Kansas City Chiefs", and a column for the date when each data point was collected. 

Now, we will create 3 new columns to consolidate all dates in each year and all dates in each month. This step will help simplify creating plots over different time periods later. 

```python
search_copy = pd.DataFrame()
search_copy = search_data_combined
def year_only(string):
    if '2014' in string:
        output = '2014'
    elif '2015' in string:
        output = '2015'
    elif '2016' in string:
        output = '2016'
    elif '2017' in string:
        output = '2017'
    elif '2018' in string:
        output = '2018'
    elif '2019' in string:
        output = '2019'
    elif '2020' in string:
        output = '2020'
    elif '2021' in string:
        output = '2021'
    elif '2022' in string:
        output = '2022'
    elif '2023' in string:
        output = '2023'
    elif '2024' in string:
        output = '2024'
    else:
        output = np.nan
    
    return output

def month_only(string):
    if '-01-' in string:
        output = '01'
    elif '-02-' in string:
        output = '02'
    elif '-03-' in string:
        output = '03'
    elif '-04-' in string:
        output = '04'
    elif '-05-' in string:
        output = '05'
    elif '-06-' in string:
        output = '06'
    elif '-07-' in string:
        output = '07'
    elif '-08-' in string:
        output = '08'
    elif '-09-' in string:
        output = '09'
    elif '-10-' in string:
        output = '10'
    elif '-11-' in string:
        output = '11'
    elif '-12-' in string:
        output = '12'
    else:
        output = np.nan
    
    return output

search_copy["Year Only"] = search_copy['Time'].apply(year_only)
search_copy['Month'] = search_copy['Time'].apply(month_only)
search_copy['Month/Year'] = search_copy['Month'] + ['/'] + search_copy['Year Only']
search_copy.head()
```

#### Check for Stationarity

For the ARIMA model to be applicable and make accurate predictions, we need to ensure our data is stationary. A stationary time series data is one whose properties do not depend on the time. We use the Adfuller test to check if the data is stationary or not. For this test, the null hypothesis will be that our data is non-stationary, while the alternative hypothesis will be that our data is stationary.

```python
from statsmodels.tsa.stattools import adfuller

dftest = adfuller(weekly_averaged_viewers['Viewers'])
print("1. ADF : ", dftest[0])
print("2. P-Value : ", dftest[1])
print("3. Num of Lags : ", dftest[2])
print("4. Num of Observations Used for ADF Regression and Critical Values Calculation : ", dftest[3])
print("5. Critical Values : ")
for key, val in dftest[4].items():
    print("\t", key, ": ", val)
```

We can see that we have a p-value that is less than 0.05. Therefore, we can reject our null hypothesis and that our data is stationary. Therefore we can proceed with our ARIMA modeling.

### Seasonal ARIMA (SARIMA)

We decided to look into using a SARIMA model which takes into account seasonality. Since our data consists of NFL viewership by weeks per season, we see a seasonal pattern in terms of viewership where towards the end of each season, the viewership sharply increases; this is likely due to the playoffs and SuperBowl. SARIMA may be a better fit for our data since these annual changes in viewership can negatively affect the original ARIMA prediction. 

Our first step is to split the dataset again into the training and testing dataset. However, we made an observation that the 2015 dataset is incomplete as it stops at week 17. Therefore, we decided on only using data from the 2016 season onwards to forecast for the 2023 season.

```python
train2 = weekly_averaged_viewers.iloc[38:-21]
test2 = weekly_averaged_viewers.iloc[-21:]
plt.plot(train2["Viewers"], color='green', label='Training')
plt.plot(test2['Viewers'], color='blue', label='Testing')
plt.title('Weekly Average Viewership Training and Test Data')
plt.xlabel('Week')
plt.ylabel('Viewers')
plt.legend()
```

The m value for the auto arima fitting is the number of periods after which we expect a pattern to repeat. Since we are using the data for an NFL season and there are typically 21 weeks in an NFL season, I am using 21 for the m value. This makes sense since every 21 weeks of NFL season, the SuperBowl happens and we see the viewership number increase sharply.

```python
results2 = auto_arima(y=train2['Viewers'], m=21)
predictions = pd.Series(results2.predict(n_periods = len(test)))
# print(predictions)
```

```python
train2['Viewers'].plot(legend=True, color='green', figsize=(12, 6), label='Training')
test2['Viewers'].plot(legend=True, color='blue', figsize=(12, 6), label='Testing')
predictions.plot(legend=True, color='red', figsize=(12, 6), label='Prediction')
plt.title('Weekly Average Viewership Training, Test, and Prediction Data')
plt.xlabel('Week')
plt.ylabel('Viewers')
```

Here we see that the viewership prediction is much more closely related than our initial ARIMA prediction. Also, the prediction line actually predicts the next 21 weeks (ie a full season)! We can take a closer look comparing the training and testing data.

```python
test2['Viewers'].plot(legend=True, color='blue', figsize=(12, 6), label='Testing')
predictions.plot(legend=True, color='red', figsize=(12, 6), label='Prediction')
plt.title('Weekly Average Viewership Test and Prediction Data')
plt.xlabel('Week')
plt.ylabel('Viewers')
```

We can see that the prediction is slightly lower than the actual data for the observed (testing) viewership but nicely follows the trend we would expect to see for a season of viewership for the NFL. Interperting this result, we see that there is not a significant difference between the forecasted and real results. Since the SARIMA model is trained on data from prior to Taylor Swift and Travis Kelce began dating, the forecasted data is what would be the viewership trend had Taylor Swift and Travis Kelce never dated. Therefore, the data suggests that there wasn't a huge impact from Taylor Swift and Travis Kelce's relationship on the overall 2023 TV viewership trend.

```python
from sklearn.metrics import r2_score
test2['predicted_viewers'] = predictions
r2_score(test2['Viewers'], test2['predicted_viewers'])
```

From our predictions, we calculated a r-squared value, which tells us "the proportion of the variance in the dependent variable that is predictable from the independent variable(s)." [Wikipedia](https://en.wikipedia.org/wiki/Coefficient_of_determination) This means that around 88% of the variability in the actual viewership data is explained by the model. However, it also implies that around 12% of the variability remains unexplained by the model. This unexplained variability could be due to random factors and confounds that uniquely affect each season in our data.

A better test is to the Ljung-Box Test (LB-Q) which checks if autocorrelation exists within a time series. Autocorrelation means "the degree of correlation of the same variables between two successive time intervals." [Source](https://corporatefinanceinstitute.com/resources/data-science/autocorrelation/) This is a good fit for our model since our predicted model is using data without Taylor Swift and Travis Kelce and correlating it against our test data to see if there is a correlation. For this test, our **Null Hypothesis** is: The residuals are independently distributed. Our **Alternate Hypothesis** is: The residuals are not independently distributed; they exhibit serial correlation. Based on what we see from the r-squared value, we would expect to fail to reject the null hypothesis, meaning that we expect a p-value greater than 0.05.

```python
import statsmodels.api as sm
#First calculate residuals from predicted vs. test
test2['resid'] = test2['predicted_viewers'] - test2['Viewers']
sm.stats.acorr_ljungbox(test2['resid'], lags=[20], return_df=True)
```

As we can see 0.13 > 0.05 meaning that we fail to reject the null hypothesis. This suggests that our residuals are independently distributed, so we cannot correlate the predicted data and the test data. This suggests, as we suspected with the r-squared value, that Taylor Swift and Travis Kelce's relationship didn't have a significant impact on average viewership for all NFL games.
